```python
import numpy as np
import matplotlib.pyplot as plt

# 
np.random.seed(0)
x1 = np.array([0.1,0.15,0.2,0.25,0.3,0.35])
y1 = np.array([0.4,0.5,0.4,0.3,0.45,0.35])

# 
plt.figure(figsize=(20, 15),dpi=400)

# 
plt.scatter(x1, y1, facecolors='w', edgecolors='darkblue', marker='o', linewidth=4, s=600)

x2 = x1+0.3
y2 = y1+0.4

# 
plt.scatter(x2, y2, facecolors='w', edgecolors='darkred', marker='o', linewidth=4, s=600)


x3 = x2+0.35
y3 = y2-0.2

# 
plt.scatter(x3, y3, facecolors='w', edgecolors='darkgreen', marker='o', linewidth=4, s=600)

x4 = x3+0.2
y4 = y3-0.4

# 
plt.scatter(x4, y4, facecolors='w', edgecolors='saddlebrown', marker='o', linewidth=4, s=600)

# 
plt.xlim(0.05, 1.3)
plt.ylim(0, 1)
plt.axis('off')
plt.grid(alpha=0.7)

# plt.xticks(fontsize=0.1)
plt.tick_params(axis='both', which='major', labelsize=30)  # 


# 
plt.show()

plt.savefig('1.png', transparent=True)
```
